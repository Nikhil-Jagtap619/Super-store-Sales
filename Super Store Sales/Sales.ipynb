## Seaborn practice

#### seaborn is a python library based on matplotlib provides high level interface for drawing and informative statistical graphics

### Sales data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sales = pd.read_excel("sales.xlsx")
sales.head(7)
```

### Relational Plot


```python
plt.figure(figsize=(8,4), dpi=120)
sns.relplot(x="Sales", y="Profit", data=sales, hue="Customer Segment", style="Ship Mode")
plt.title("Sales & Profit")
plt.show()
```

### Line Chart

```python
sales["Year"] = sales["Order Date"].dt.year
# sales["Year"].unique()


sns.relplot(x="Year", y="Sales",data=sales, kind="line", hue="Ship Mode",col="Product Category", style="Customer Segment", ci=None)
```

### Categorical Plot

#### Scatter Chart

```python
# we'll have the same dataset as used above
sns.catplot(x="Product Category", y="Sales", data=sales, col="Customer Segment", col_wrap=2, hue="Ship Mode")
```

### Swarm Chart

```python
# sns.catplot(x="Product Category", y="Sales", data=sales, kind="swarm")
```

### Box Plot Chart

```python
sales.head(2)
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind="box", hue="Ship Mode", col="Ship Mode")
```

### Violin Plot 

```python
sns.catplot(x="Product Category", y="Shipping Cost", data=sales, kind="violin", col="Ship Mode")
```

### Bar Plot

```python

sns.catplot(x="Product Category", y="Sales", data=sales, kind="bar", hue="Product Container", col="Ship Mode")
```

### Point Plot

```python
sns.catplot(x='Product Category', y='Sales', data=sales, kind='point',hue="Ship Mode")
```
